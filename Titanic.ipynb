```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#link do danych: https://uploads.kodilla.com/bootcamp/ds/10/titanic.csv
df = pd.read_csv('titanic.csv')
df.head()
```

```python
sns.heatmap(df[['Fare']].isnull())
```

```python
print(df['Fare'].shape)
```

```python
mean_fare = np.round(np.mean(df['Fare']))
mean_fare
```

```python
median_fare = np.round(np.median(df['Fare']))
median_fare
```

```python
plt.hist(df['Fare'], bins=20)
plt.axvline(x = mean_fare, color='pink', label = 'rednia')
plt.axvline(x = median_fare, color='salmon', label = 'Mediana')
plt.legend(loc='upper right')
plt.show()
```

```python
q0 = np.quantile(df['Fare'], 0.0)
q1 = np.round(np.quantile(df['Fare'], 0.25))
q3 = np.round(np.quantile(df['Fare'], 0.75))
q4 = np.round(np.quantile(df['Fare'], 1.0))
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_fare}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
```

```python
plt.boxplot(df['Fare'])
plt.show()
```

```python
variance_fare = np.round(np.var(df['Fare']))
variance_fare
```

```python
standard_deviation_fare = np.round(np.std(df['Fare']))
standard_deviation_fare
```

```python
sns.displot(df['Fare'], kind="ecdf")
```
